```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

inside_data = pd.read_csv('inside_at_2-50pm_2-24-23.csv')
display(inside_data)
```

```python
# defining normal distribution curve function
def normal_dist(x , mean , sd):
    prob_density = np.exp(-0.5*((x-mean)/sd)**2)/(math.sqrt(2*np.pi)*sd)
    return prob_density
```

```python
# PM10

x_pm10_in = inside_data["PM10"]

mean_pm10_in = np.mean(x_pm10_in)
std_pm10_in = np.std(x_pm10_in)

# x_norm = np.linspace(np.min(x),np.max(x),200)
# norm_dist = normal_dist(x_norm,mean,std)

plt.hist(x_pm10_in,bins=1, density=True)
# plt.plot(x_norm,norm_dist) cant plot normal distribution curve because std is zero
plt.show
```

```python
# PM25

x_pm25_in = inside_data["PM25"]

mean_pm25_in = np.mean(x_pm25_in)
std_pm25_in = np.std(x_pm25_in)

x_norm_pm25_in = np.linspace(np.min(x_pm25_in),np.max(x_pm25_in),200)
norm_dist_pm25_in = normal_dist(x_norm_pm25_in,mean_pm25_in,std_pm25_in)

plt.hist(x_pm25_in,bins=2,density=True)
plt.plot(x_norm_pm25_in,norm_dist_pm25_in)
plt.show
```

```python
# PM100

x_pm100_in = inside_data["PM100"]

mean_pm100_in = np.mean(x_pm100_in)
std_pm100_in = np.std(x_pm100_in)

x_norm_pm100_in = np.linspace(np.min(x_pm100_in),np.max(x_pm100_in),200)
norm_dist_pm100_in = normal_dist(x_norm_pm100_in,mean_pm100_in,std_pm100_in)

plt.hist(x_pm100_in, bins=2, density=True)
plt.plot(x_norm_pm100_in,norm_dist_pm100_in)
plt.show
```

- Get data from another group in your lab and compare your inside data with their inside data, including all of elements from Step 4. Do the same comparing your outside data with their outside data.
- How do the sensors compare? What do you think this indicates about the accuracy and precision of your sensors?

The differences in the sensors' means and stds will tell us something about their relative accuracy and precision. If the means are relatively close then it means that they are relatively accurate. If their stds are relatively close then it shows that the precision of the sensor is relatively good. We can't know how accurate a specific sensor is unless we compare it to a sensor with a known accuracy, however we can tell that the precision is good by just how small the standard deviation is. We compared the sensors' accuracy and precision using the differences in the means and stds for the temperature to the altitude measurements. We only discussed the accuracy and precision of the sensors for the inside temperature measurements to the inside altitude measurements as it was very repetitive and feel that we got the point across for how the means and stds say something about the accuracy and precision.
